```python
import pandas as pd
food_df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding = 'latin-1')
print(food_df.shape) #To obtain no of rows and columns in the data set
food_df.head() #dataset preview
```

```python
#To find the total number of unique countries in the dataset
food_df['Area'].nunique()
```

```python
#To find the total protein supply quantity in Madagascar in 2015
Mada =food_df[food_df['Area'] =='Madagascar'] #
Mada
```

```python
#To find the total protein supply quantity in Madagascar in 2015
mada_protein = Mada[Mada['Element'] =='Protein supply quantity (g/capita/day)']
mada_protein['Y2015'].sum()
```

```python
#To find the standard deviation and mean across the entire dataset for 2017
food_df.describe()
```

```python
#Total number of missing data in 2014
food_df['Y2014'].isnull().sum()
```

```python
#Finding which year had the loest correlation with 'Element Code'
food_df.corr()
```

```python
#To find which 'Area' had the '7th' Lowest sum in 2017
food_df_area_and_y2017 = food_df[['Area', 'Y2017']]
sum_food = food_df_area_and_y2017.groupby('Area')['Y2017'].sum()
sum_food.sort_values()
```

```python
#To find which 'Area' had the Highest sum in 2017
sum_food.sort_values(ascending=False)
```
